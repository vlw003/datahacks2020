```python
def drop_hotspot(x):
    hotspots = ["Fisherman's Wharf, San Francisco, CA",'Oracle Park, 24 Willie Mays Plaza, San Francisco, CA',\
                'The Palace Of Fine Arts, 3601 Lyon St, San Francisco, CA']
    if x in hotspots:
        return x
    else:
        return np.nan
```

```python
# there are two fishermans wharf names. lambda function makes it into one.
bhot['Destination Display Name'] = bhot.apply(lambda x: "Fisherman's Wharf, San Francisco, CA" if x['Destination Display Name'] == "Fisherman's Wharf, 286-298 Jefferson St, San Francisco, CA" else x['Destination Display Name'],axis=1)
bhot_copy = bhot.copy()
# bhot_stat = bhot.groupby(['Origin Display Name', 'Destination Display Name']).count()[['Daily Mean Travel Time (Seconds)']]
# bhot_stat = bhot_stat.reset_index()
bhot_copy['Origin Display Name'] = bhot_copy['Origin Display Name'].apply(drop_hotspot)
bhot_nonan = bhot_copy.dropna(subset = ['Origin Display Name'])
bhot_nonan.groupby(['Origin Display Name', 'Destination Display Name']).mean()[['Daily Mean Travel Time (Seconds)']]
# gives us rough estimate of fastest bart stop from hotspot, numbers arent accurate (mean of means)
```

**This function classifys the military times as early morning, midday, evening**

```python
def time_convert(x):
    if x>=0 and x<7:
        return 'early morning'
    elif x>=7 and x<15:
        return 'midday'
    else:
        return 'evening'
```

**Groupby military time and travel(origin and destination code combined). This is to get the means that will later be hard coded into a function that fills np.nan**

```python
h1_fromBart['hod'] = h1_fromBart['hod'].apply(time_convert)
h1_fromBart.groupby(['hod', 'travel']).mean()[['mean_travel_time']]
# make a graph vs time and vs location
```

```python
h2_fromBart['hod'] = h2_fromBart['hod'].apply(time_convert)
h2_fromBart.groupby(['hod', 'travel']).mean()[['mean_travel_time']]
```

```python
def date_convert(x):
    to_str = x[:2]
    to_int = int(to_str)
    if to_int <=3:
        return 'q1'
    else:
        return 'q2'
```

```python
clean = bh.copy()
clean['quarter'] = clean['Date'].apply(date_convert)
clean = clean[['Date', 'travel','quarter','Early Morning Mean Travel Time (Seconds)','Evening Mean Travel Time (Seconds)','Midday Mean Travel Time (Seconds)']]
clean.head()
```

**Helper functions to convert nans into early, evening, midday respectively**

```python
def gen_early(x):
    if pd.isnull(x):
        return 'early'
    else:
        return x

def gen_evening(x):
    if pd.isnull(x):
        return 'evening'
    else:
        return x

def gen_midday(x):
    if pd.isnull(x):
        return 'midday'
    else:
        return x
```

**Apply functions and combine columns: travel, quarter, and respective time column (early morning, evening, midday) into their own column**

```python
lst =["36033792","36923792","37603792"]
clean = clean.loc[clean['travel'].isin(lst)]
clean['Early Morning Mean Travel Time (Seconds)'] = clean['Early Morning Mean Travel Time (Seconds)'].apply(gen_early)
clean['Evening Mean Travel Time (Seconds)'] = clean['Evening Mean Travel Time (Seconds)'].apply(gen_evening)
clean['Midday Mean Travel Time (Seconds)'] = clean['Midday Mean Travel Time (Seconds)'].apply(gen_midday)
clean['test early']=clean['travel'].astype(str)+clean['quarter'].astype(str)+clean['Early Morning Mean Travel Time (Seconds)'].astype(str)
clean['test evening']=clean['travel'].astype(str)+clean['quarter'].astype(str)+clean['Evening Mean Travel Time (Seconds)'].astype(str)
clean['test midday']=clean['travel'].astype(str)+clean['quarter'].astype(str)+clean['Midday Mean Travel Time (Seconds)'].astype(str)
clean
```

**Helper function that returns the mean to each unique code, if the value was never nan, return the original value**

```python
def damn(x):
    if x == '36033792q2early':
        return 419.498571
    elif x =='36923792q2early':
        return 413.530000
    elif x == '37603792q2early':
        return 591.407143
    elif x == '36033792q1early':
        return 444.078571
    elif x == '36923792q1early':
        return 404.667143
    elif x == '37603792q1early':
        return 606.072857
    elif x == '37603792q2evening':
        return 761.708889
    elif x == '36033792q1evening':
        return 619.613333
    elif x == '36923792q1evening':
        return 501.261111
    elif x == '37603792q1evening':
        return 803.496667
    elif x == '36923792q2midday':
        return 473.085000
    elif x == '37603792q2midday':
        return 773.660000
    elif x == '37603792q1midday':
        return 858.106250
    else:
        return float(x[-5:])
```
